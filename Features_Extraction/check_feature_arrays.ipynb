```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_5 = np.load(r'D:\Hanze\Master\Python_Course\test\Data_processing\5_sec_feature_array.npz')
data_10 = np.load(r'D:\Hanze\Master\Python_Course\test\Data_processing\10_min_feature_array.npz')
```

```python
X5 = data_5['arr_0']
X10 = data_10['arr_0']
```

```python
print(f'The shape of 5 seconds features {X5.shape}')
print(f'The shape of 10 minutes features {X10.shape}')
```

```python
print(f'The first row of X5: {X5[1]}')
print(f'The first row of X10: {X10[0,:,1]}')
```

```python
# Create a figure with 7 rows and 6 columns
fig, axes = plt.subplots(7, 6, figsize=(15, 12))

# Plot each of the 37 columns
for i, ax in enumerate(axes.flat):
    if i < 37:
        ax.plot(X5[:, i])  # Plot the i-th column of the dataset
        ax.set_title(f"Feature {i+1}")
        ax.set_xticks([])  # Optional: remove x ticks for cleaner plots
        # ax.set_yticks([])  # Optional: remove y ticks for cleaner plots
        print(f'Feature {i+1}:')
        print(f'--  Min = {np.min(X5[:, i])}')
        print(f'--  Max = {np.max(X5[:, i])}')
        print(f'--  Diff = {np.max(X5[:, i]) - np.min(X5[:, i])}')

    else:
        ax.axis("off")  # Turn off the empty subplots

plt.tight_layout()
plt.show()
```
