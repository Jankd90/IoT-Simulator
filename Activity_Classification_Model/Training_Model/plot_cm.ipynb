```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
df_cm_10 = pd.read_csv(r'confusion_matrix_acm_10_row_percentage.csv', index_col=0)
df_cm_10_meta = pd.read_csv(r'confusion_matrix_acm_10_meta_row_percentage.csv', index_col=0)
```

```python
# Function to plot the confusion matrix
def plot_confusion_matrix(cm, classes):
    plt.figure(figsize=(13, 10))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix (%)')
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)

    # Print the numbers in each cell of the matrix
    threshold = cm.max() / 2.  # Threshold for coloring
    for i, j in np.ndindex(cm.shape):
        plt.text(j, i, format(cm[i, j], '.2f'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > threshold else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    plt.show()
```

```python
# Convert DataFrame to numpy array
cm_10 = df_cm_10.values  
classes_10 = df_cm_10.index.to_list()  # Extract class names
# cm, classes

# Plot the confusion matrix
plot_confusion_matrix(cm_10, classes_10)
```

```python
# Convert DataFrame to numpy array
cm_10_meta = df_cm_10_meta.values  
classes_10_meta = df_cm_10_meta.index.to_list()  # Extract class names
# cm, classes

# Plot the confusion matrix
plot_confusion_matrix(cm_10_meta, classes_10_meta)
```
